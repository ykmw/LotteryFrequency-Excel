```python
import pandas as pd

# Read the Excel file into a DataFrame
df = pd.read_excel("YourExcelFile.xlsx")

# Calculate the frequency of each number in Ball 1
ball1_freq = df['Ball 1'].value_counts()
ball1_numbers = ball1_freq.index.tolist()
ball1_frequencies = ball1_freq.tolist()

# Calculate the frequency of each number in Ball 2
ball2_freq = df['Ball 2'].value_counts()
ball2_numbers = ball2_freq.index.tolist()
ball2_frequencies = ball2_freq.tolist()

# Calculate the frequency of each number in Ball 3
ball3_freq = df['Ball 3'].value_counts()
ball3_numbers = ball3_freq.index.tolist()
ball3_frequencies = ball3_freq.tolist()

# Calculate the frequency of each number in Ball 4
ball4_freq = df['Ball 4'].value_counts()
ball4_numbers = ball4_freq.index.tolist()
ball4_frequencies = ball4_freq.tolist()

# Calculate the frequency of each number in Ball 5
ball5_freq = df['Ball 5'].value_counts()
ball5_numbers = ball5_freq.index.tolist()
ball5_frequencies = ball5_freq.tolist()

# Create a new DataFrame with labeled columns
data = {
    'Ball 1': ball1_numbers,
    'Ball 1 Frequencies': ball1_frequencies,
    'Ball 2': ball2_numbers,
    'Ball 2 Frequencies': ball2_frequencies,
    'Ball 3': ball3_numbers,
    'Ball 3 Frequencies': ball3_frequencies,
    'Ball 4': ball4_numbers,
    'Ball 4 Frequencies': ball4_frequencies,
    'Ball 5': ball5_numbers,
    'Ball 5 Frequencies': ball5_frequencies,
}

max_length = max(map(len, data.values()))  # Find the maximum length

# Pad the lists with None to make them of equal length
data = {key: value + [None] * (max_length - len(value)) for key, value in data.items()}

# Create the DataFrame
frequencies_df = pd.DataFrame(data)

# Export the DataFrame to an Excel file
frequencies_df.to_excel("YourExportFile.xlsx", index=False)

# Print a confirmation message
print("Excel file 'YourExportFile.xlsx' has been created.")
```
